```python
#import matplotlib

import matplotlib.pyplot as plt
import pandas as pd 
plt.style.use("fast")      
%matplotlib notebook

# Import data
data = pd.read_csv("data.csv")

data
```

### Visualization Story
I am creating different charts to analyze how the temperature affected sales of Dept 1, Dept 2 & Dept 3 respectively.



### 1) Histogram

```python
# This histogram plot reviews the frequency of the different temperature ranges.

data.plot.hist(y = 'Temperature')


# It can be seen that temperatures of 80 - 85 occurred 28 times.
```

### 2) Bar Chart

```python
# The Bar Chart below is comparing the sales of Dept 1, Dept 2 and Dept 3.
# I displayed the first 10 rows to keep the visualization clean.

data[:10].plot.bar(y = ["dept_1 sales", "dept_2 sales", "dept_3 sales"], x = 'Date')
```

### 3) Line Chart

```python
# The Line Chart below is comparing the sales of Dept 1, Dept 2 and Dept 3.

data.plot(y = ["dept_1 sales", "dept_2 sales", "dept_3 sales"], x = "Date")
```

### 4) Area Chart

```python
# The Area Chart below is reviewing the sales of Dept 1, Dept 2 and Dept 3.

data.plot.area(x = "Date", y = ["dept_1 sales", "dept_2 sales", "dept_3 sales"])
```

### 5) Scatter Plot

```python
# The Scatter plot below is looking at the Department 2 sales in respect to Temperature.

data.plot.scatter(x = 'Temperature', y = 'dept_2 sales')
```
