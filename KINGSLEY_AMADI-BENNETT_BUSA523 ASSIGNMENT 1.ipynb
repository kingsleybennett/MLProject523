```python
# Question 1 Solution:

import random

guesses = 0

print("Hi, Welcome to the Guessing game. What is your name?")
Name1 = input()

print("Well, " + Name1 + ", I am thinking between 1 and 20. Can you guess the number?")


num = random.randint(1, 20)

while guesses < 5:

     print("Take a guess...") 
     Myguess = input()
     Myguess = int(Myguess)
     guesses = guesses + 1

     if Myguess != num:
        print("Your guess is incorrect.")
        
     if Myguess < 1 or Myguess > 20:
        print("Please input a valid number between 1 and 20") #Line to determine if the input is </> than the range
     elif Myguess == num:
         break


if Myguess == num:   
     guesses = str(guesses)
     print("You are correct, " + Name1 + "! You guessed the number in " + guesses + " guesses!")
elif Myguess != num:
     num = str(num)
     print("You did not guess correctly. The number I was thinking of was " + num)
```
